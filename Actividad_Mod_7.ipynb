```python
import pandas as pd
iris = pd.read_csv("iris.csv")
iris.head()
```

1. Determinar en una sola tabla lo siguiente:
Media, desviación y mediana para el ancho del petalo por cada tipo de flor para aquellos registros cuyo ancho de sépalo es mayor o igual a 3.5 y el largo del sepalo es mayor que 5

```python
#Primero, filtramos registros cuyo ancho de sépalo sea mayor o igual que 3.5 y el largo del sepalo mayor que 5

registros = iris[(iris['sepal.width'] >= 3.5) & (iris['sepal.length'] > 5.0)]                                               
registros
```

```python
#Agrupamos por tipo de flor y calculamos la media, desviación y mediana para el ancho del petalo.

estadisticos = registros.groupby('variety')['petal.width'].agg(['mean','std','median'])
estadisticos
```

2. Calcular el número de registros que se utilizaron en el ejercicio anterior

```python
num_registros = len(registros)
print(f'Se utilizaron {num_registros} registros')
```

3. Agregar una nueva columna al dataframe generado al problema 1, que sea el resultado de estandarizar los valores del ancho del petalo para cada registro. Denomine dicha columna "petal.width.st" y calcule de la siguiente manera: a cada observación de la columna "petal.width.st" se le resta su media de la columna para luego dividir entre la desv

```python
# Estandarizar los valores del ancho del pétalo para cada registro
media = registros['petal.width'].mean()
desv = registros['petal.width'].std()

registros['petal.width.st'] = (registros['petal.width'] - media) / desv
registros

```

4. Mantenga el dataframe del problema 3 "variety" y "petal.with.st" y renombrar como "variedad" y "ancho.est.petalo"

```python
registros.rename(columns={"variety": "variedad", "petal.width.st": "ancho.est.petalo"}, inplace=True)
registros
```

5.Exportar el resultado del data frama a un archivo nuevo llamado iris2.csv

```python
registros.to_csv("iris2.csv", index=False)
```

6. Generar un gráfico de dispersión para la base original Iris que relacione la variable "petal.width" con "petal.length"

```python
import matplotlib.pyplot as plt
plt.scatter(iris["petal.width"], iris["petal.length"])

# añadir títulos y etiquetas de los ejes
plt.title("Relación entre el ancho y largo del pétalo")
plt.xlabel("Ancho del pétalo")
plt.ylabel("Largo del pétalo")

# mostrar el gráfico
plt.show()
```
